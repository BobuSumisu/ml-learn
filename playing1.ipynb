# Playing 1: Linear Regression

Playing around with data for ex1. 

```octave
addpath("./functions");

ex1data1 = load("data/ex1data1.txt");
ex1data1 = sortrows(ex1data1);

% Take 10 rows to test with later.
data = ex1data1(1:86, :);
control = ex1data1(87:97, :);

n = columns(data);
m = rows(data);
X = [ ones(m, 1), data(:, 1) ];
y = data(:, 2);

plot(data(:, 1), data(:, 2), 'rx');
hold on;

% Linear hypothesis.

theta = gradientDescent(zeros(2, 1), X, y, 0.01, 10000);
fprintf("linear hypothesis, gradient descent: %f\n", cost(theta, X, y));
plot(data(:, 1), hypothesis(theta, X), 'r;linear, gradient descent;');

theta = normalEquation(X, y);
fprintf("linear hypothesis, normal equation: %f\n", cost(theta, X, y));
plot(data(:, 1), hypothesis(theta, X), 'r;linear, normal equation;');

% Let's try a quadratic hypothesis.

X_quad = [ ones(m, 1), data(:, 1), data(:, 1).^2 ];

theta_quad = gradientDescent(zeros(3, 1), X_quad, y, 0.0001, 10000);
fprintf("quadratic hypothesis, gradient descent: %f\n", cost(theta_quad, X_quad, y));
plot(data(:, 1), hypothesis(theta_quad, X_quad), 'g;quad, gradient descent;');

theta_quad = normalEquation(X_quad, y);
fprintf("quadratic hypothesis, normal equation: %f\n", cost(theta_quad, X_quad, y));
plot(data(:, 1), hypothesis(theta_quad, X_quad), 'g;quad, normal equation;');

% Cubic?
X_cube = [ ones(m, 1), data(:, 1), data(:, 1).^2, data(:, 1).^3 ];

theta_cube = gradientDescent(zeros(4, 1), X_cube, y, 0.000001, 10000);
fprintf("cubic hypothesis, gradient descent: %f\n", cost(theta_cube, X_cube, y));
plot(data(:, 1), hypothesis(theta_cube, X_cube), 'b;cubic, gradient descent;');

theta_cube = normalEquation(X_cube, y);
fprintf("cubic hypothesis, normal equation: %f\n", cost(theta_cube, X_cube, y));
plot(data(:, 1), hypothesis(theta_cube, X_cube), 'b;cube, normal equation;');

% Maybe a square root function.

X_sqrt = [ ones(m, 1), data(:, 1), sqrt(data(:, 1)) ];

theta_sqrt = gradientDescent(zeros(3, 1), X_sqrt, y, 0.001, 10000);
fprintf("square root hypothesis, gradient descent: %f\n", cost(theta_sqrt, X_sqrt, y));
plot(data(:, 1), hypothesis(theta_sqrt, X_sqrt), 'k;square root, gradient descent;');

theta_sqrt = normalEquation(X_sqrt, y);
fprintf("square root hypothesis, normal equation: %f\n", cost(theta_sqrt, X_sqrt, y));
plot(data(:, 1), hypothesis(theta_sqrt, X_sqrt), 'k;square root, normal equation;');
```
